## Rad sa kolonama kod dataframe-ova

### Pristup kolonama

```python
movies.MOVIES
```

```python
short_df = movies[['MOVIES','YEAR','GENRE']]
short_df.head()
```

```python
list_of_columns = ['MOVIES','RATING']
shorter_df = movies[list_of_columns]
shorter_df.head()
```

### Dodvanje novih kolona

```python
movies["type"] = "movie\tv show"
```

```python
movies.head()
```

```python
movies["NewRunTime"] = movies["RunTime"]*10
movies.head()
```

```python
movies["NEW_GENRE"] = movies["GENRE"].str.strip()
movies.head()
```

### Promena postojece kolone

```python
movies["ONE-LINE"] = movies["ONE-LINE"].str.strip()
movies["STARS"] = movies["STARS"].str.strip()
movies["STARS"] = movies["STARS"].str.replace('\n',' ')
movies["GENRE"] = movies["GENRE"].str.strip()
movies.head()
```

### Promena naziva kolone

```python
movies = movies.rename(columns={'MOVIES': 'Name', 'YEAR': 'Year', 'GENRE': 'Genre', 'STARS': 'Stars', 'VOTES': 'Votes', 'ONE-LINE':'Description','RATING':'Rating'})
movies.head()
```

### Brisanje kolona

```python
selected_columns_movies = movies[['Name','Year','Genre','Rating','Description','Stars','Votes','RunTime','Gross']]
selected_columns_movies.head()
```

```python
movies = movies.drop(columns=['type', 'NewRunTime','NEW_GENRE'], axis=1)
```

```python
movies.head()
```

## Sortiranje dataframe-ova

```python
movies.sort_values(by=["Rating", "RunTime"], ascending=[False, True])
```

```python
movies.sort_values("Rating")
movies.sort_values(by="Rating")
movies.sort_values(by="Rating", ascending=True)
movies.sort_values(by="Rating", ascending=False)

movies.sort_values("Year")
movies.sort_values("Year", ascending=False)
movies.sort_values("Year", na_position="last")
movies.sort_values("Year", na_position="first")
movies.sort_values("Year", na_position="first", ascending=False)


movies.sort_values(["Rating", "Votes"])
movies.sort_values(by=["Rating", "Votes"])
movies.sort_values(by=["Rating", "Votes"], ascending=True)
movies.sort_values(by=["Rating", "Votes"], ascending=False)
movies.sort_values(by=["Rating", "Votes"], ascending=[True, False])

movies = movies.sort_values(["Rating", "Votes"], ascending=[False, True])
movies
```
