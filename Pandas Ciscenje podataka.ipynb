```python
def lokacija(latitude,longitude):
    from geopy.geocoders import Nominatim
    geolocator = Nominatim(user_agent="data analysis course")
    location = geolocator.reverse(latitude+","+longitude)
    address = location.raw['address']
    return address.get('postcode')
```

```python
for index,row in airbnb[airbnb['zipcode'].isna()].iterrows():
    airbnb.at[index,'zipcode'] = lokacija(str(row['latitude']),str(row['longitude']))
```
