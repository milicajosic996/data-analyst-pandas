## MERGE (JOIN)

```python
pd.set_option("display.max_rows",None)
```

```python
import pandas as pd
```

```python
studenti = pd.read_excel("Fakultet.xlsx", sheet_name = "Studenti", index_col = 'student_id')
profesori = pd.read_excel("Fakultet.xlsx", sheet_name = "Profesori", index_col = 'profesor_id')
predmeti = pd.read_excel("Fakultet.xlsx", sheet_name = "Predmeti", index_col = 'predmet_id')
ocene = pd.read_excel("Fakultet.xlsx", sheet_name = "Ocene", index_col = 'ocena_id')

```

```python
studenti.head()
```

```python
studenti_ocene = studenti.merge(ocene, how = 'inner', on = "student_id")
```

```python
studenti_ocene
```

```python
studenti_ocene = studenti.merge(ocene, how = 'left', on = "student_id")
```

```python
studenti_ocene
```

```python
stu_oce_pred = studenti_ocene.merge(predmeti, how = 'inner', on = 'predmet_id')
```

```python
stu_oce_pred
```

```python
studenti_ocene = studenti.merge(ocene, how = 'right', on = "student_id")
```

```python
studenti_ocene
```

```python
studenti_ocene = studenti.merge(ocene, how = 'outer', on = "student_id")
```

```python
studenti_ocene
```

```python
studenti_ocene = studenti.merge(ocene, how = 'cross')
#studenti_ocene
```

```python
predmet_profesor = predmeti.merge(profesori, how= 'left', left_on = 'id_profesora', right_on = 'profesor_id')
predmet_profesor
```

## Pivot

```python
prodavnica = pd.read_excel("Prodavnica.xlsx")
prodavnica.head()
```

```python
pivot_table = prodavnica.pivot(index = 'Proizvod', columns = 'Tip prodaje', values = 'Prodaja')
```

```python
pivot_table
```

```python
pivot_table= pivot_table.rename_axis(None, axis=1).reset_index()
pivot_table
```

```python
pivot_table.info()
```

```python
pivot_table['Total'] = pivot_table['Online']+pivot_table['U prodavnici']
pivot_table
```

```python
melt_table = pd.melt(pivot_table, id_vars = 'Proizvod', value_vars = ["Online","U prodavnici","Total"], var_name = "Tip prodaje",value_name = "Cena")
```
