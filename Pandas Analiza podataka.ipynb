# Group By

```python
data = {
    'Prodavnica': ['A','A','A','A','A','A','B','B','B','B','B','B'],
    'Proizvod': ['Jabuka', 'Jabuka', 'Banana', 'Banana', 'Pomorandža', 'Pomorandža','Jabuka', 'Jabuka', 'Banana', 'Banana', 'Pomorandža', 'Pomorandža'],
    'Mesec': ['Januar', 'Februar', 'Januar', 'Februar', 'Januar', 'Februar','Januar', 'Februar', 'Januar', 'Februar', 'Januar', 'Februar'],
    'Prodaja': [100, 120, 80, 90, 70, 80, 110, 70, 130, 80,100, 90],
    'Trošak': [50, 60, 40, 45, 30, 35, 55, 50, 45, 40, 30, 40]
}
df = pd.DataFrame(data)
df
```

```python
groups = df.groupby('Prodavnica')
groups
```

```python
df.groupby('Prodavnica').get_group('A')
```

```python
df.groupby('Prodavnica')['Prodaja'].mean()
```

```python
df.groupby('Prodavnica')['Prodaja'].min()
```

```python
df.groupby('Prodavnica')['Prodaja'].max()
```

```python
df.groupby('Prodavnica')['Prodaja'].sum()
```

```python
df.groupby('Prodavnica')['Prodaja'].count()
```

```python
df.groupby(['Prodavnica','Mesec'])['Trošak'].mean()
```

```python
df.groupby(['Prodavnica','Mesec']).mean()
```

```python
df.groupby(['Prodavnica','Mesec']).mean().round(2).reset_index()
```

```python
group = airbnb.groupby("neighbourhood")["price"].mean()
```

```python
group.sort_values()
```

```python
group = airbnb.groupby("room_type")["price"].mean()
group
```

```python
group = airbnb.groupby(['room_type','neighbourhood'])['price'].mean()
```

```python
group
```

```python
group = airbnb.groupby(['room_type','neighbourhood']).agg({'price':'mean'}).sort_values(['room_type','price'])
group
```

```python
group = airbnb.groupby(['room_type','neighbourhood']).agg(min_price = ('price','min'),max_price = ('price','max')).sort_values(['room_type'])
group
```

```python
superhost_stats = airbnb.groupby(['room_type','host_is_superhost'])[['price','review_scores_rating']].mean()
superhost_stats
```

## Titanik

```python
titanik = pd.read_csv('titanik_clean.csv')
titanik.head()
```

```python
columns_to_explore = ['Survived','Pclass','Sex','Embarked']
for column in columns_to_explore:
    print(titanik[column].value_counts(normalize = True)*100)
```

```python
titanik.describe()
```

```python
cat_col = titanik.select_dtypes(include = 'number')
```

```python
cat_col.corr()
```

```python
ct_titanic = pd.crosstab(index = titanik['Survived'], columns = titanik['Sex'], margins = True)
ct_titanic.index = ['nije preziveo','preziveo','ukupno']
ct_titanic
```

```python
ct_titanic = pd.crosstab( index = titanik['Survived'], columns = titanik['Pclass'], margins = True)
ct_titanic.index = ['nije preziveo','preziveo','ukupno']
ct_titanic.columns = ['Prva','Druga','Treca','ukupno']
ct_titanic
```

```python
ct_titanic = pd.crosstab( index = titanik['Survived'], columns = [titanik['Pclass'],titanik['Sex']], margins = True)
ct_titanic.index = ['nije preziveo','preziveo','ukupno']
ct_titanic
```
