## Dodavanje novih kolona

```python
filmovi.head()
```

## Promena postojece kolone

```python
filmovi['ONE-LINE'] = filmovi['ONE-LINE'].str.strip()
```

```python
filmovi['STARS'] = filmovi['STARS'].str.strip()
```

```python
filmovi.head()
```

```python
filmovi['STARS'] = filmovi['STARS'].str.replace('\n', ' ')
filmovi.head()
```

```python
filmovi['GENRE'] = filmovi['GENRE'].str.replace('\n', ' ')
filmovi.head()
```
