## Apply funkcija nad kolonom Dataframe-a

```python
def komentari(row):
    price = row.loc['price']
    room_type = row.loc['room_type']
    
    if price<100:
        return 'Cheap'
    if room_type == 'Entire home/apt':
        return 'Entire place for yourself'
    return 'No comment'
```

```python
airbnb['comment'] = airbnb.apply(komentari, axis = 'columns')
```

```python
airbnb[['price','room_type','comment']].head(50)
```
